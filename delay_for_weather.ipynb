```python
import pandas as pd
import numpy as np
import itertools
```

```python
FARES_DATA = r"fares_with_distance_fixed.csv"
fares_data = pd.read_csv(FARES_DATA)
```

# Load weather data and add columns for day/month/year

```python
WEATHER_DATA = r"C:\Users\ASUS\OneDrive\Semester 3\Innovation Project\Assign_02\weatherAUS\weatherAUS.csv"
```

```python
weather_data = pd.read_csv(WEATHER_DATA)
```

```python
weather_data["Date_Convert"] = pd.to_datetime(weather_data["Date"])
```

```python
weather_data["Month"] = weather_data["Date_Convert"].map(lambda x : x.month)
weather_data["Year"] = weather_data["Date_Convert"].map(lambda x : x.year)
weather_data["Day"] = weather_data["Date_Convert"].map(lambda x : x.day)
weather_data
```

# Resample weather data to have a entry for each month rather than each day
To do this, we will use ``pd.resample()`` and set each month's columns to the average of all days within the month

```python
resampled_weather_data = weather_data.resample(rule='M', on='Date_Convert').mean()
resampled_weather_data
```

```python
resampled_weather_data["Month"] = resampled_weather_data.index.map(lambda x: x.month)
resampled_weather_data["Year"] = resampled_weather_data.index.map(lambda x: x.year)
resampled_weather_data
```

```python
resampled_weather_data = resampled_weather_data.interpolate(method='polynomial', order=5)
```

```python
resampled_weather_data
```

```python
fares_data
```

```python
fares_with_weather = pd.merge(
    left=fares_data,
    right=resampled_weather_data,
    how='outer',
    left_on=['Month', 'Year'],
    right_on=['Month', 'Year'],
)
```
